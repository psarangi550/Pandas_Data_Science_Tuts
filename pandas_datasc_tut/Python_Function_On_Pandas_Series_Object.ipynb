# <ins>Python Core function By Passing the Series Object</ins>

    - len():- we can pass a pandas Series Object which will return the length of that specific `Pandas Series Object`
    - type():- we can pass the `Pandas Series Object` into the inbuild `type()` which will return the type of Specific Type Object into it
            :- when we pass the `Pandas Series Object` then it will tell what `type of values` stored inside the `Pandas Series Object`
            
    - dir() :- by using the `dir()` and passing the `Pandas Data Series` we can fetch the `available attribute and function inside of it`
    - sorted() :- by using the `sorted()` we can sort the values of the pandas Series object
               :- But this `sorted()` will create a `New Series object which will be already sorted out`
               :- this will provide the `Pandas Series Object Values` in a sorted format
    
    - reversed() :- will reverse the order of values in the `Pandas Series object`
                 :- we can get the `reverse sorted value`   by usign the command as `sorted(<pandas Series obj>,reversed=True)`
                 
    - list()  :- we can convert the `Pandas Series object` into the `python list` by using the `list()`
    - dict()  :- we can convert the `Pandas Series object` into the `Python dict` by using the dict() where the key is the index and values of Pandas Series                                                          become the value
              :- But the `pandas Series Object` can have duplicated Keys  while Turning them into the dict we need to make sure the index were not duplicated
              
              
    - max()   :- return the `maximum value in case of string(last Letter Words)` as the single value from the `Series Object`
    - min()   :- return the `minimum value in case of string(First Letter Words)` as the single value from the `Series Object`
    
    

```python
import pandas as pd
#importing the Pandas Module in here
df=pd.read_csv("datasets/google_stock_price.csv", usecols=["Stock Price"])
#creating the DataFrame object from the read_csv() and making a single column by using the params as usecols
google=df.squeeze()
#converting the DataFrame into Series Object in this case
len(google)
#fetching the length of the Series Object
type(google)
#fetching the attribbute and methods from the dir()
dir(google)
```
